```python
#categorizing games by time increments - rapid, blitz, bullet, classical
def game_category(val):
    if val <= 3:
        return 'bullet'
    elif val <= 10:
        return 'blitz'
    elif val < 60:
        return 'rapid'
    elif val >= 60:
        return 'classical'
    else:
        return 'other'
```

```python
df['game_category'] = df['game_time'].apply(game_category)
df['game_category'].value_counts()
```
